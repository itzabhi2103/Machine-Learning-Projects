# Importing the **libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('online_payment_fraud.csv')
data.head()
```

# checking the NAN **values**

```python
data.isnull().sum()
```

# counting the type of **transactions**

```python
data.type.value_counts()
```

```python
type = data['type'].value_counts()
transactions = type.index
quantity = type.values
```

# Checking there percentage

```python
import plotly.express as px
figure = px.pie(data,
                    values = quantity,
                    names = transactions,
                    hole = 0.5,
                    title = "Distribution of Transaction Type")
figure.show()
```

```python
numeric_data = data.select_dtypes(include=np.number)
numeric_data
```

# implementing the correlation between **isFraud**

```python
correlation = numeric_data.corr()
print(correlation['isFraud'].sort_values(ascending=False))
```

# giving the categorical name to the type of transactions

```python
data["type"] = data["type"].map({"CASH_OUT": 1, "PAYMENT": 2,
                                 "CASH_IN": 3, "TRANSFER": 4,
                                 "DEBIT": 5})
data["isFraud"] = data["isFraud"].map({0: "No Fraud", 1: "Fraud"})
data.head()
```

# Online Fraud Detectation Model

```python
x = np.array(data[['type', 'amount', 'oldbalanceOrg', 'newbalanceOrig']])
y = np.array(data[['isFraud']])
```

splitting the data into test case and training case

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.10, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(xtrain, ytrain)
print(model.score(xtest, ytest))
```

# Now testing the model by feeding inputes

```python
features = np.array([[4, 9800, 9800, 0]])
print(model.predict(features))
```
