# Cleaning the **data**

```python
# creating a backup copy
backup = data.copy()
```

```python
# opting only data which are numeric
data = data[data['year'].str.isnumeric()]
```

```python
# changing year object to integer
data['year'] = data['year'].astype(int)
```

```python
# exclusing the Ask for price term
data = data[data['Price']!="Ask For Price"]
```

```python
# price converted into integer and excluded ','
data['Price'] = data['Price'].apply(str).str.replace(',','').astype(int)
```

```python
data
```

```python
data.info()
```

```python
# spliting data and taking only the first elements excluding kms
data['kms_driven'] = data['kms_driven'].str.split(' ').str.get(0)
```

```python
data.info()
```

```python
# excluding , with '' and converting str into integer
data['kms_driven'] = data['kms_driven'].str.replace(',','').astype(int)
```

```python
data.info()
```

```python
data['kms_driven']
```

```python
data = data[data['kms_driven'].apply(str).str.isnumeric()]
```

```python
data
```

```python
# excluding the null row
data = data[~data['fuel_type'].isna()]
```

```python
# taking first three name from name
data['name'] = data['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
data
```

```python
# Reseting the index
data = data.reset_index(drop=True)
```

```python
data
```

```python
# removing the outliers we will accept those cars whose values are below 60 lakhs
data = data[data['Price']<6e6].reset_index(drop=True)
```

```python
data.describe()
```

```python
data.to_csv('Cleaned Car.csv')
```

```python
x = data.drop(columns='Price')
y = data['Price']
```

# feature scaling

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

# linear **regression**

```python
from sklearn.linear_model import LinearRegression
```

# Creating an OneHotEncoder object to contain all the possible categories

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore')
ohe.fit(x[['name', 'company', 'fuel_type']])
```

# Creating a column transformer to transform categorical columns

```python
from sklearn.compose import make_column_transformer
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                    remainder='passthrough')
```

# Linear Rgeression Model

```python
lr = LinearRegression()
```

# Making a pipeline

```python
from sklearn.pipeline import make_pipeline
pipe = make_pipeline(column_trans, lr)
```

# Fitting the model

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

#Checking R2 Score

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

# Making Predictions

```python
pipe.predict(pd.DataFrame(columns=X_test.columns,data=np.array(['Maruti Suzuki Swift','Maruti',2019,100,'Petrol']).reshape(1,5)))
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```
